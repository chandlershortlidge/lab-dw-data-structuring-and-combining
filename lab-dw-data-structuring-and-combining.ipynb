1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# pivot_df = df.pivot_table(index='sales_channel', columns='year', values=['GDP'], aggfunc='sum')
pivot_df = ca_clean.pivot_table(index="policy", columns="sales_channel", values=["customer_lifetime_value"], aggfunc='sum')
pivot_df
```

```python
pivot_df.sum()

# using sum, we can clearluy see that agents make the most sales across all sales channels, followed by branch, call center, and finally, online.
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_df2 = ca_clean.pivot_table(index="gender", columns="education", values=["customer_lifetime_value"], aggfunc='mean')
pivot_df2

# very low education (HS or below) & very high education (Master) result in the highest average CLV for men and women. 
# otherwise, the variation doesn't produce very much meaning across genders. HS F produce somewhat more than HS M, but otherwise results are quite similar.
```
